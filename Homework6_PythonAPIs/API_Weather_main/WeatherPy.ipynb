```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).



```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"
query_url = f"{url}appid={api_key}&units={units}&q="

cities_list = []
cloudiness = []
country = []
date = []
humidity = []
lat= []
lng = []
max_temp = []
wind_speed = []

print("Beginning Data Retrieval \n-----------------------------")

counter = 1
for city in cities:
    query = query_url + city
    try:
        response = requests.get(query).json()
        print(f'Processing record {counter} of 500. | {city}')
        cities_list.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp'])
        wind_speed.append(response['wind']['speed'])
        counter += 1
    except:
        print("City not found. Skipping...")
    if counter == 500:
        break
        
print("----------------------------- \nData Retrieval Complete \n -----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Check that all lists have same length
print(f'City {len(cities_list)}')
print(f'Length {len(cloudiness)}')
print(f'Country {len(country)}')
print(f'Date {len(date)}')
print(f'Humidity {len(humidity)}')
print(f'Latitude {len(lat)}')
print(f'Longitude {len(lng)}')
print(f'Max temperature {len(max_temp)}')
print(f'Wind speed {len(wind_speed)}')
```

```python
city_weather = pd.DataFrame({'City' : cities_list, 'Cloudiness':cloudiness, 'Country':country, "Date":date,
                            'Humidity':humidity, "Lat":lat, 'Lng':lng, "Max Temp":max_temp, "Wind Speed":wind_speed})
#Export city_weather dataframe to csv
city_weather.to_csv("../Output/cities.csv")

date_mean = city_weather['Date'].mean()
date_mean = time.strftime("%m/%d/%y", time.gmtime(date_mean))
city_weather.head()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Temperature vs Latitude Plot

```python
plt.figure(1)
plt.scatter(city_weather['Lat'], city_weather['Max Temp'], alpha=0.7, edgecolors = 'k')
plt.title(f'Max Temperature vs Latitude {date_mean}')
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
plt.grid(True)
```

#### Humidity vs Latitude Plot

```python
plt.figure(2)
plt.scatter(city_weather['Lat'], city_weather['Humidity'], color= 'orange', alpha=0.7, edgecolors = 'k')
plt.title(f'Humidity vs Latitude {date_mean}')
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid(True)
```

#### Latitude vs. Cloudiness Plot

```python
plt.figure(3)
plt.scatter(city_weather['Lat'], city_weather['Cloudiness'], color= 'green', alpha=0.7, edgecolors = 'k')
plt.title(f'Cloudiness vs Latitude {date_mean}')
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid(True)
```

#### Wind Speed vs Latitude Plot

```python
plt.figure(4)
plt.scatter(city_weather['Lat'], city_weather['Wind Speed'], color= 'red', alpha=0.7, edgecolors = 'k')
plt.title(f'Humidity vs Latitude {date_mean}')
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid(True)
```

## Analysis


* From the Max Temperature vs Latititude plot, we can see that the highest temperatures are clustured within 20 degrees of the equator (0 degrees) and that max temperatures decrease the with increasing distance from the equator.

* From the Cloudiness vs Latititude plot, cloudiness is uniformly distributed across latitutdes; there is no relationship between cloudiness and latitude.

* From the Wind Speed vs Latitude plot, wind speed is uniformly distributed across latitudes; there is no relationship between wind speed and latitude.

* From our randomly generated latitutde and longitude coordinates, more cities are in the Northern Hemisphere than in the Southern Hemisphere. Resultantly, our data is skewed to the right.
